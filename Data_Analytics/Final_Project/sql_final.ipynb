#### Задание 2  
Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
query_2 = '''
select 
    distinct 
    book_id, 
    count(review_id) as rvw_cnt,
    avg(rating) as avg_rating
from books left join reviews using(book_id)
           left join ratings using(book_id)
group by 1

'''
```

```python
pd.io.sql.read_sql(query_2, con = engine) 
```

#### Задание 3  
Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
query_3 = '''
    with stg1 as (
    select 
        publisher, 
        count(book_id)
    from publishers inner join books using(publisher_id)
    where num_pages > 50
    group by 1
    order by 2 desc
    )

    select publisher  from stg1
    limit 1
'''
```

```python
pd.io.sql.read_sql(query_3, con = engine) 
```

#### Задание 4  
Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
query_4 = '''
    with stg1 as (
        select 
            author,
            avg(rating) as rating, 
            count(rating_id) as rating_cnt
        from books inner join authors using(author_id)
                   inner join ratings using(book_id)
    group by 1
    having count(rating_id) >50 
    order by 2 desc
    )

    select author from stg1
    limit 1
'''
```

```python
pd.io.sql.read_sql(query_4, con = engine) 
```

#### Задание 5  
Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query_5 = ''' 
    with stg1 as (
    select
        r.username,
        count(review_id) as review_cnt
    from ratings r inner join reviews using(book_id) 
    group by 1
    having count(rating_id) > 50
    )

    select avg(review_cnt) as avg_reviews from stg1
'''
```

```python
pd.io.sql.read_sql(query_5, con = engine) 
```

Среднее количество обзоров на пользователя - `164`
